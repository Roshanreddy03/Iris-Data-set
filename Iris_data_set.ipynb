Load the Dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

iris_df = sns.load_dataset('iris')

iris_df.head()
```

1. Check the Balance of the Target Variable

```python
# Count the occurrences of each species
species_counts = iris_df['species'].value_counts()
print("Species Count:\n", species_counts)
```

2. Create Histograms and Box Plots

Histogram

```python
iris_df.hist(figsize=(10, 8), bins=15, edgecolor='black')
plt.suptitle('Histograms of Iris Features')
plt.show()
```

Box plot

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris_df.drop(columns=['species']))
plt.title('Box Plots of Iris Features')
plt.show()
```

3. Visualize Relationships with a Pairplot

```python
# Create a pairplot color-coded by species
sns.pairplot(iris_df, hue='species', markers=["o", "s", "D"])
plt.suptitle('Pairplot of Iris Dataset', y=1.02)
plt.show()
```

4. Compute Correlation Matrix and Heatmap

```python
numerical_df = iris_df.select_dtypes(include='number')

# Compute the correlation matrix
corr_matrix = numerical_df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Iris Features')
plt.show()
```

Summary

- Distinct Separation: Iris setosa is highly distinct from the other two species. Looking at the pairplot, setosa forms a completely isolated cluster, primarily characterized by having much smaller petal lengths and petal widths.

- Overlap: Iris versicolor and Iris virginica share similar sepal dimensions and have some overlap in petal sizes. However, virginica generally has larger petal lengths and widths compared to versicolor.

- Feature Correlation: The heatmap reveals a very strong positive correlation between petal length and petal width. Sepal length also shows a strong positive correlation with both petal metrics, while sepal width appears mostly uncorrelated or slightly negatively correlated with the other features.
